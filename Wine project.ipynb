 importing train test split to split the data and train it 

```python
from sklearn.model_selection import train_test_split
x_train, x_test , y_train , y_test = train_test_split(x,y,test_size=0.30, random_state=9)
```

# using Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
```

IMPORTING ACCURACY SCORE AND CONFUSION  MATRIX == 
 to check the accuracy of the model getting in each algorithm

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy_score(y_test,y_pred)
```

# Decision Tree:

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier(criterion='entropy',random_state=9)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
print("Accuracy store:", accuracy_score(y_test,y_pred))
```

# KNN:

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model=KNeighborsClassifier(n_neighbors=3)
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
```
